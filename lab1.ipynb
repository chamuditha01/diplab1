```python
import cv2
import matplotlib.pyplot as plt

# Assuming bw_image2 and image_details are properly defined images
# Calculate histograms
name_for_histogram = cv2.calcHist([bw_image2], [0], None, [256], [0, 256])
name_for_histogram1 = cv2.calcHist([image_details], [0], None, [256], [0, 256])

# Plot the histograms
plt.plot(name_for_histogram, label='bw_image2 Histogram')
plt.plot(name_for_histogram1, label='image_details Histogram')

# Set title and labels once
plt.title('Image Histograms')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.legend()  # Add a legend to distinguish the two histograms

# Show the plot
plt.show()
```
