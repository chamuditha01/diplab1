```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "/content/Lenna_(test_image).png"
imglena = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(imglena)
```

```python
imglena.shape
```

```python
print(imglena)
```

```python
# Define upsampling factor
scale_factor = 2
# Get original image dimensions
rows, cols = imglena.shape
print(rows,cols)
# Calculate new image dimensions
```

```python
# Calculate new image dimensions
scale_factor = 2
new_rows, new_cols = imglena.shape[0] * scale_factor, imglena.shape[1] * scale_factor
print(new_rows, new_cols)
```

```python
resized_img = cv2.resize(imglena, (new_cols, new_rows),
interpolation=cv2.INTER_NEAREST)
```

```python
# Display original and interpolated images
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.imshow(imglena, cmap='gray'), plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(resized_img, cmap='gray'), plt.title('Nearest Neighbor Interpolation')
```

```python
plt.show()
# Display original and resized images
cv2_imshow(imglena)
cv2_imshow(resized_img)
```

```python
imglenabilinear = cv2.resize(resized_img, (new_cols, new_rows),
interpolation=cv2.INTER_LINEAR)
```

```python

# Display both interpolated images
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.imshow(resized_img, cmap='gray'), plt.title('Nearest Neighbor Interpolation')
plt.subplot(1,2,2)
plt.imshow(imglenabilinear, cmap='gray'), plt.title('Bilinear Interpolation')
plt.show()
```

```python
plt.show()
# Display original and resized images
cv2_imshow(imglenabilinear)
cv2_imshow(resized_img)
```

```python

lenahistogram=cv2.calcHist([resized_img],[0],None,[255],[0,255])
plt.plot(lenahistogram)
plt.title('intensity representation')
plt.xlabel('pixel')
plt.ylabel('count')
plt.grid(True)
plt.show()
```

```python
negativeimg = 255 - imglena
cv2_imshow(imglena)
cv2_imshow(negativeimg)
```

```python
hist_orig = cv2.calcHist([imglena], [0], None, [256], [0, 256])
hist_negative = cv2.calcHist([negativeimg], [0], None, [256], [0, 256])

# Plot both histograms in separate subplots
plt.figure(figsize=(12, 6))

# Plot original image histogram
plt.subplot(1, 2, 1)
plt.plot(hist_orig, color='blue')
plt.title('Original Image Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.grid(True)

# Plot negative image histogram
plt.subplot(1, 2, 2)
plt.plot(hist_negative, color='green')
plt.title('Negative Image Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.grid(True)

plt.show()
```

```python
c = -1 #scaling constant
loglena = c * np.log(1 + imglena)
plt.imshow(loglena, cmap='gray')
plt.show()
```

```python
hist_orignal = cv2.calcHist([imglena], [0], None, [256], [0, 256])
hist_log = cv2.calcHist([loglena], [0], None, [256], [0, 256])

# Plot both histograms in separate subplots
plt.figure(figsize=(12, 6))

# Plot original image histogram
plt.subplot(1, 2, 1)
plt.plot(hist_orignal, color='blue')
plt.title('Original Image Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.grid(True)

# Plot log-transformed image histogram
plt.subplot(1, 2, 2)
plt.plot(hist_log, color='green')
plt.title('Log-Transformed Image Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.grid(True)

plt.show()
```
