```python
# Choose the desired kernel size (e.g., 3 for a 3x3 kernel)
kernel_size = 15
# Apply box filtering using cv2. filter2D with a normalized box kernet
filtered_img = cv2.filter2D(lena, -1, np.ones((kernel_size, kernel_size), np.float32) / (kernel_size**2))
filtered_img2 = cv2. filter2D(lena, -1, np.ones ((23, 23), np.float32) / (kernel_size**2))

cv2_imshow(np.hstack((lena,filtered_img,filtered_img2)))
```

```python
kernal_size=(5,5)
sigma=100.0

kernel = cv2.getGaussianKernel(kernal_size[0],sigma)

filtered_img3 = cv2.filter2D(lena,-1,kernel)

cv2_imshow(np.hstack((lena,filtered_img3)))
```

```python
# Sobel filter in x-direction (detects vertical edg
xaxis = cv2. Sobel(lena, -1, 1, 0, ksize=3)
# Sobel filter in y-direction (detects horizontal edges)
yaxis = cv2. Sobel(lena, -1, 0, 1, ksize=3)
# combine results using:
name_for_x_combined_sobel= xaxis + xaxis
final_out = name_for_x_combined_sobel
# -1 for default depth, ksizevis kernel

cv2_imshow(np.hstack((lena, final_out)))
```
