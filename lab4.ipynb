```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Optional for visualization
```

```python
# Optional for visualization
from google.colab. patches import cv2_imshow
# Convert original to grayscale
lenna_gray_img = cv2. imread ('/content/Lenna_(test_image).png', cv2. IMREAD_GRAYSCALE)
laplacian = cv2. Laplacian (lenna_gray_img, cv2.CV_8U)
# If you want to detect both edges, better option is to keep the output datatype to some higher forms, like cv.CV_165, cv. CV_64F, cv.8U etc
cv2_imshow(np.hstack( (lenna_gray_img, laplacian)))
```

```python
#Custom Kernel
laplacian_kernel = np.array ([
[-1, -1, -1],
[-1, 8, -1],
[-1, -1, -1]])
#apply to image
manually_filtered_image = cv2.filter2D(lenna_gray_img, -1, laplacian_kernel)
#Depth of the output image I -1 will give the output image depth as same as the input imagel
cv2_imshow(np.hstack((lenna_gray_img, laplacian,manually_filtered_image) ))
```
